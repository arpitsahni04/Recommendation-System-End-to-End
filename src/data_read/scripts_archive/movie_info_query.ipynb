## all the files

```python
def get_movie_info(res, mid):
    line = [str(int(mid)), res['id'], str(res['tmdb_id']), res['imdb_id'], '\"'+res['title']+'\"', '\"'+res['original_title']+'\"', res['adult'], res['budget']]

    genres = ["%d:%s"%(item['id'],item['name']) for item in res['genres']]
    line.append("|".join(genres))

    line.extend([res['original_language'], res['popularity']])

    countries = ["%s:%s"%(item['iso_3166_1'],item['name']) for item in res['production_countries']]
    line.append("|".join(countries))

    line.extend([res['release_date'], res['revenue'], str(res['runtime'])])

    langs = ["%s:%s"%(item['iso_639_1'],item['name']) for item in res['spoken_languages']]
    line.append("|".join(langs))

    line.extend([res['status'], res['vote_average'], res['vote_count']])

    return ",".join(line) + "\n"
```

```python
header = "movie_id,movie_code,tmdb_id,imdb_id,title,original_title,adult,budget,genres,original_language,popularity,production_countries,release_date,revenue,runtime,spoken_languages,status,vote_average,vote_count\n"

movie_info = open("processed_data_v2/movie_info_v2.csv", "w")
movie_info.write(header)

f = open("processed_data_v2/movie_id_v2.csv", "r")
f.readline()

l = f.readline()
i = 0
while l:
    mcode = l.split(",")[0]
    mid = l.split(",")[1]
    res = requests.get("http://128.2.204.215:8080/movie/%s" %mcode)
    if res.status_code != 200:
        print(mcode, res.json())
        l = f.readline()
        continue

    res = res.json()
    line = get_movie_info(res, mid)
    movie_info.write(line)
    l = f.readline()

    i += 1
    if i % 100 == 0:
        print(i)

movie_info.close()
```
